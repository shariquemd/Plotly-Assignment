```python
Q1. Scatter plot for age and fare columns in the titanic dataset:

python
Copy code
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Plot scatter plot using Plotly express
scatter_plot = px.scatter(titanic, x='age', y='fare', title='Scatter Plot: Age vs Fare')
scatter_plot.show()
Q2. Box plot for the tips dataset:

python
Copy code
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot box plot using Plotly express
box_plot = px.box(tips, x='day', y='total_bill', title='Box Plot: Total Bill by Day')
box_plot.show()
Q3. Histogram for "sex" and "total_bill" columns in the tips dataset:

python
Copy code
import plotly.express as px

# Plot histogram using Plotly express
histogram_plot = px.histogram(tips, x='sex', y='total_bill', color='smoker', facet_col='day',
                               title='Histogram: Total Bill by Sex and Smoker')
histogram_plot.show()
Q4. Scatter matrix plot for the iris dataset:

python
Copy code
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Plot scatter matrix using Plotly express
scatter_matrix = px.scatter_matrix(iris, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                                   color='species', title='Scatter Matrix Plot: Iris Dataset')
scatter_matrix.show()
Q5. Distplot using Plotly express:

python
Copy code
import plotly.express as px

# Distplot using Plotly express
distplot = px.histogram(tips, x='total_bill', marginal='rug', nbins=30, title='Distplot: Total Bill Distribution')
distplot.show()
A Distplot is a histogram combined with a kernel density estimate. It shows the distribution of a univariate set of observations and provides a smoothed representation of the distribution. The marginal='rug' parameter adds a rug plot along the x-axis to show individual data points.
```
